```python
import pandas as pd
```

```python
df_train=pd.read_csv("train.csv")
df_test=pd.read_csv("test.csv")
```

```python
df_train
```

```python
df_test
```

```python
import preprocessing_pipeline_app
```

```python
preprocessor=preprocessing_pipeline_app.custom_preprocessing()
```

```python
df_train['preprocessed_text']=df_train['Text_data'].apply(preprocessor.preprocess)
```

```python
x_train=preprocessor.vectorize(df_train['preprocessed_text'])
```

```python
pd.DataFrame(x_train.toarray(), columns=preprocessor.vectorizer.get_feature_names_out())
```

```python
df_train
```

```python
df_test['preprocessed_text']=df_test['text data'].apply(preprocessor.preprocess)
```

```python
df_test
```

```python
x_test=preprocessor.transform(df_test['preprocessed_text'])
```

```python
y_test=df_test['Class labels']
```

```python
y_train=df_train['Label']
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(multi_class="multinomial", solver="lbfgs", max_iter=1000)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
model = LogisticRegression(multi_class="multinomial", solver="lbfgs", max_iter=1000, class_weight="balanced")
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.svm import LinearSVC
model = LinearSVC(class_weight="balanced")
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
##LinearSVC is the best model so far. need other non skewed datasets
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(ngram_range=(1,3), max_features=20000, min_df=3)
```

```python
x_train=vectorizer.fit_transform(df_train['preprocessed_text'])
```

```python
x_test=vectorizer.transform(df_test['preprocessed_text'])
```

```python
model = LinearSVC(class_weight="balanced")
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))## best so far
```

```python
from sklearn.calibration import CalibratedClassifierCV
```

```python
svc = LinearSVC(class_weight="balanced")
model = CalibratedClassifierCV(svc, cv=5)  
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
