```python
import pandas as pd 
```

```python
df=pd.read_csv("train.csv")
```

```python
df.drop(df.keys()[0],axis=1,inplace=True)
```

```python
df.isnull().any()
```

```python
import re
```

```python
df['Text_data']=df['Text_data'].apply(lambda x: re.sub(r"[^a-zA-Z\s]","",x))
```

```python
df['Text_data']=df['Text_data'].apply(lambda x: re.sub(r'\b\w*(\w)\1{2,}\w*\b', '',x).lower())
```

```python
df
```

```python
from nltk.tokenize import word_tokenize
```

```python
df['Text_data']=df['Text_data'].apply(lambda x: word_tokenize(x))
```

```python
df
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words("english"))
```

```python
df['Text_data']=df['Text_data'].apply(lambda x:[w for w in x if w not in stop_words])
```

```python
df
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
from nltk.corpus import wordnet
```

```python
pos_map={
    'J':wordnet.ADJ,
    'N':wordnet.NOUN,
    'V': wordnet.VERB,
    'R':wordnet.ADV
}
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
def lemmatize_tokens(tokens):
    pos_tags=nltk.pos_tag(tokens)
    return [lemmatizer.lemmatize(word,pos_map.get(tag[0],wordnet.NOUN)) for word,tag in pos_tags]
```

```python
df['Text_data']=df['Text_data'].apply(lemmatize_tokens)
```
